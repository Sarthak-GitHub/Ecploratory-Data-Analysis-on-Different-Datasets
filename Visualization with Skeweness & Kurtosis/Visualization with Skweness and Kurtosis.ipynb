## Visualization with Skweness and Kurtosis

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Create new random.normal(size=100000) and visualize mean and median with the density plot. Also, note your observations.

```python
import random
thread = np.random.normal(size=100000)
sns.distplot(thread,hist=False,kde_kws={'shade':True},color='seagreen')
plt.axvline(np.mean(thread),label='Mean',color='orange')
plt.axvline(np.median(thread),label='Median',color='yellow')
plt.legend()
plt.show()
```

**Observations**
- The obtained distribution is symmetrical (S ~= 0)(skeweness)
- The mean and median of data lies in center of distribution 
- The distribution is mesokurtic (kurtosis)

#### Create new skewed_data random.exponential(size=100000) and visualize mean and median with the density plot. Also, note your observations. 

```python
skewed_data = np.random.exponential(size=100000)
sns.distplot(skewed_data,hist=False,kde_kws={'shade':True},color='seagreen')
plt.axvline(np.mean(skewed_data),label='Mean',color='red')
plt.axvline(np.median(skewed_data),label='Median',color='orange')
plt.legend()
plt.show()
```

**Observations**
- The given distribution is right/positively skewed as the (S>0)
- mean < median tells that the data is right skewed

#### Create data and outliers data using random.normal(size=50) and visualize mean and median with the density plot. Also, note your observations.

```python
data = np.random.normal(size=50)
outliers = np.random.normal(loc=10, size=50)

club = np.concatenate((data,outliers))

sns.distplot(club,hist=False,kde_kws={'shade':True},color='seagreen')
plt.axvline(np.mean(club),label='Mean',color='red')
plt.axvline(np.median(club),label='Median',color='green')

plt.legend()
plt.show()
```

**Observations**
- The data is bimodal and the mean as well as median for this 
- Distribution of data is approximately near to each other.

## Skeweness and Kurtosis

Beyond measures of center and spread, descriptive statistics include measures that give you a sense of the shape of a distribution. Skewness measures the skew or asymmetry of a distribution while kurtosis measures the "peakedness" of a distribution.

#### To explore these two measures further, let's first create some dummy data (norm_data, skewed_data, uniform_data, peaked_data) and inspect it ?

```python
norm_data = np.random.normal(size=35000)
skewed_data = np.concatenate((norm_data+2,np.random.exponential(size=65000)),axis=0)
uniform_data = np.random.uniform(0,2, size=100000)
peaked_data = np.concatenate((np.random.exponential(size=50000),np.random.exponential(size=50000)*(-1)),axis=0)

fig, ax= plt.subplots(2,2,figsize=(18, 10))
sns.distplot(norm_data,hist=False,kde_kws={'shade':True},color='b',ax=ax[0,0])
sns.distplot(skewed_data,hist=False,kde_kws={'shade':True},color='r',ax=ax[0,1])
sns.distplot(uniform_data,hist=False,kde_kws={'shade':True},color='g',ax=ax[1,0])
sns.distplot(peaked_data,hist=False,kde_kws={'shade':True},color='orange',ax=ax[1,1])
ax[0,0].set_title('Normal Distribution')
ax[0,1].set_title('Skewed Distribution')
ax[1,0].set_title('Uniform Distribution')
ax[1,1].set_title('Peaked Distribution')
plt.show()
```

**Observations**
- norm_data is symmetrically distributed  (mesokurtic)
- skewed _data is right/positively skewed
- uniform_data is platykurtic in nature
- peaked_data is leptokurtic

```python
import pandas as pd
from pandas import Series

norm_data = Series(norm_data)
skewed_data = Series(skewed_data)
uniform_data = Series(uniform_data)
peaked_data = Series(peaked_data)

print("Skeweness of norm_data is :", norm_data.skew())
print("Skeweness of skewed_data is :", skewed_data.skew())
print("Skeweness of uniform_data is ", uniform_data.skew())
print("Skeweness of peaked_data is :", peaked_data.skew())
print('='*50)
print("Kurtosis of norm_data is :", norm_data.kurt())
print("Kurtosis of skewed_data is :", skewed_data.kurt())
print("Kurtosis of uniform_data is :", uniform_data.kurt())
print("Kurtosis of peaked_data is :", peaked_data.kurt())
```

**SKEWENESS**

- If the skeweness is S ~= 0 then, the the distribution is symmetrical.
- If the skeweness is S > 0 then, the the distribution is right/positively skewed.
- If the skeweness is S < 0 then, the the distribution is left/negatively skewed.
- In the above readings, the values for norm_data, uniform_data and peaked_data are near to 0 so, we can assume it as symmetrical and the value for skewed data is above 0 and greater than 1, that is clearly a right skewed distribution.

**KURTOSIS**

- If the kurtosis is around zero then, it is mesokurtic.
- If the kurtosis is greater than 3 then, it is leptokurtic
- n above readings, norm_data is mesokurtic (value near to zero)
- the skewed data is kind of leptopkurtic
- the uniform data is platykurtic (value less than 0)
- the peaked data is leptokurtic (value greater than 3)
